```python
from load_llm import run_vqa

# This is an example of how you can run LLMs
# You need their API keys first, so please fill in the apis.py
models = ['claude-sonnet-4-5', 'claude-haiku-4-5', 'gpt-5', 'gemini-2.5-pro']
answers = {}

for model in models:
    assert model in models, f"Must be one of the following: {models}"
    answers[model] = {}
    
    image_path =f'./tree/subtree/colors_3.png'
    prompt = f"Is this an example of the subtree of another tree?" + " Return the final answer in curly brackets (e.g., {Yes} or {No})."
    answer = run_vqa(model, prompt, image_path)
    print(model, answer)
```

So here, we see that the model sees the word true and outputs true, disregarding their reasoning chain.
Here, we can say that these models are biased toward the word "True"
