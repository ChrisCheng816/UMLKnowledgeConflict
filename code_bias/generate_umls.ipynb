```python
import json

# To create UML diagrams, you need to have a json file
# Please check the files isa_relationships_json and hasa_relationship_json
# These files contains example for IsA and HasA relationships, which typically models have strong knowledge
with open('hasa_relationships_json.json', 'r') as f:
  isa_top25 = json.load(f)
```

```python
import os
for category in isa_top25:
  if not os.path.exists(f'umls/hasA/{category}'):
    os.mkdir(f'umls/hasA/{category}')

  for pairs in (isa_top25[category]):
    class_child_name = pairs['part']
    class_parent_name = pairs['owner']

    class_child =  {
        'name': class_child_name,
        'attributes': [],
        'methods': []
    }
    class_parent =  {
        'name': class_parent_name,
        'attributes': [],
        'methods': []
    }
    draw_UML(class_child, class_parent, save_path=f'umls/hasA/{category}/{class_parent_name}->{class_child_name}.png', relationship='hasa')
```
